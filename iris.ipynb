```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


iris = sns.load_dataset("iris")
```

```python
iris['species'].value_counts()
```

```python
iris.head()
```

```python
col = 'sepal_length'
sns.histplot(data=iris, x=col, kde=True).set_title(f'Distribuição da variável {col}')
```

```python


for col in iris.drop(columns='species'): #Fluxograma de variáveis contínuas a partir de uma condição para que todas as variáveis contínuas sejam contempladas
    sns.histplot(data=iris, x= col, kde= True, hue=iris['species']).set_title(f'Distribução da variável {col}')
    plt.show()
   
```

```python
for col in iris.drop(columns='species'): #Cria um grafico em média para cada variável categórica
    sns.barplot(data=iris, x=iris['species'], y= col, hue= iris['species'], ci= 90)
    plt.title('Bar plot of Iris dataset')
    plt.show()
```
